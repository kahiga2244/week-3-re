```
#cleaning Telcom_dataset3.csv
with open('cells_geo.csv','r') as f:
  df = pd.read_csv(f,index_col=0,encoding='utf',sep=';')
print(df.columns)
#renaming columns
df.rename(columns={'SITE_CODE':'SITE_ID'}, inplace=True)
#dropping columns i dont actuually need
df_new = df.drop(['STATUS','AREA','ZONENAME','LOCALISATION','LATITUDE','LONGITUDE'], axis=1)
print(df_new.head(5))
#looking for null values
df_new.isnull()
#DROPPING THE NULL VALUES
df_new_no_missing = df.dropna()
df_new_no_missing
#CLEANING
df_new_cleaned = df.dropna(how='all')
df_new_cleaned
#REPLACE WITH ZERO
df_new.fillna(0)




```

```
# MERGING THE DATASETS1, DATASETS2 and DATASETS3

combined_dataset = pd.concat([df4, df2, df3])
combined_dataset.drop_duplicates(keep='last',inplace=True)
combined_dataset
```

```
#dropping columns 
#combined_dataset.drop(['STATUS','AREA','ZONENAME','LOCALISATION','LATITUDE','LONGITUDE'], axis=1)
combined_dataset = combined_dataset.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
combined_dataset
```

```
#dropping columns 
combined_dataset.dropna()
```

```
combined_dataset.drop(['DW_A_NUMBER_INT','DW_B_NUMBER_INT','COUNTRY_A','COUNTRY_B'], axis=1)
```

```
# WE SPLIT THE DATE_TIME COLUMN TO FORM DATE AND TIME SEPATE COLUMNS

combined_dataset['DATE']= combined_dataset['DATETIME'].str.split(" ").str[0] 
combined_dataset['TIME']= combined_dataset['DATETIME'].str.split(" ").str[1]

combined_dataset.head()
```

```
#dropping columns i dont need again
combined_dataset.drop(['DW_A_NUMBER_INT','DW_B_NUMBER_INT','COUNTRY_A','COUNTRY_B'], axis=1)
```

```
#merging combined(df,df2,df3) with df_new

combined_new = pd.merge(combined_dataset,df_new,how='inner',left_on='SITE_ID',right_on='SITE_ID')
#
combined_new=combined_new.drop_duplicates()
combined_new = combined_new.rename({'VILLES':'CITY'}, axis=1)

combined_new.head()
```

```
combined_new.drop(['DW_A_NUMBER_INT','DW_B_NUMBER_INT','COUNTRY_A','COUNTRY_B'], axis=1)
```

```
#Which ones were the most used city for the three days?

most_used = combined_new['SITE_ID'].groupby(combined_new['CITY'])
most_used.count().sort_values(ascending = False)
```

```
#least used city for the three days
least_used = combined_new['SITE_ID'].groupby(combined_new['CITY'])
least_used.count().sort_values(ascending = True)
```

```
# Most used  and least used region
region = combined_new['SITE_ID'].groupby(combined_new['REGION'])
region.count().sort_values(ascending = False)
```

```
#Most used product
product = combined_new['SITE_ID'].groupby(combined_new['PRODUCT'])
product.count().sort_values(ascending = False)
```

```
#MOST USED ZONE
Zone = combined_new['SITE_ID'].groupby(combined_new['DECOUPZONE'])
Zone.count().sort_values(ascending = False)
```
